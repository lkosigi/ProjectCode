```python
images=[]
labels=[]
for sf in sub_folders:
    os.chdir(os.path.join(data_dir,'Train',sf))
    for fl in os.listdir():
        images.append(cv2.cvtColor(cv2.imread(fl),cv2.COLOR_BGR2GRAY))
        labels.append(sf)
        
```

```python
#reshape from 2D to 1D
images_flat=  []
for row in images:
    images_flat.append(row.reshape(32*32))
    
```

```python
train = pd.DataFrame(data=images_flat)
```

```python
train['labels'] = labels
```

```python
train = train.sample(frac=1).reset_index(drop=True)
```

```python
train.drop(columns='labels',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
y_train=enc.fit_transform(y_train)

```

```python
x=np.array(train)
y=keras.utils.to_categorical(y_train)
```

```python
x=x/255.0
```

```python
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from keras.layers.core import Dense,Flatten
from keras.layers.pooling import MaxPooling2D
from keras.utils import np_utils
from keras.optimizers import SGD
from keras.layers import Dropout
```

```python
model=Sequential()
model.add(Conv2D(filters=6,kernel_size=(3,3),padding='same',input_shape=(32,32,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(filters=16,kernel_size=(3,3),padding='valid'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dropout(0.2,seed=100))
model.add(Dense(120,activation='relu'))
model.add(Dense(84,activation='relu'))
model.add(Dense(46,activation='softmax'))
```

```python
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True).

















model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

```python
##Reshaping: nrows,32,32,1 channel (images are greyscale)
x_train=x.reshape(x.shape[0],32,32,1)
```

```python
history=model.fit(x_train,y,epochs=30,batch_size=1000,validation_split=0.20)
```
