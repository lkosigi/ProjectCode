```python
import os, shutil
import pandas as pd
import keras
```

```python
from keras.applications.vgg16 import VGG16,preprocess_input
from keras.preprocessing.image import load_img,img_to_array
from keras.preprocessing import image
import numpy as np
from keras.applications.vgg16 import decode_predictions
```

```python
model=VGG16()
```

```python
#https://www.programcreek.com/python/example/89223/keras.preprocessing.image.load_img
def extract_feature(dir_path, net):
    result = []

    for image_name in os.listdir(dir_path):
        image_path = os.path.join(dir_path, image_name) 
        img = image.load_img(image_path, target_size=(224, 224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        preds = decode_predictions(model.predict(x), top=1)
        result.append(preds[0][0][1])
    return result
```

```python
features = []
features = extract_feature(image_folder, model)
```

```python
def copy_image_to_dir(row): 
    img_path = os.getcwd()
    folder = ''
    if row.Image_Label == 'airliner':
        folder = 'airliner'
        img_path = img_path + r'\airliner'
        shutil.copy(row.filename,img_path)
        
    elif row.Image_Label == 'acoustic_guitar':
        folder = 'acoustic'
        img_path = img_path + r'\acoustic'
        shutil.copy(row.filename,img_path)
        
    elif row.Image_Label == 'Arabian_camel':
        folder = 'arabian'
        img_path = img_path + r'\arabian'
        shutil.copy(row.filename,img_path)
        
    elif row.Image_Label == 'Chihuahua':
        folder = 'chihuahua'
        img_path = img_path + r'\chihuahua'
        shutil.copy(row.filename,img_path)
        
    elif  row.Image_Label == 'church':
        folder = 'church'
        img_path = img_path + r'\church'
        shutil.copy(row.filename,img_path)
        
    elif  row.Image_Label == 'German_shepherd':
        folder = 'german'
        img_path = img_path + r'\german'
        shutil.copy(row.filename,img_path)
    
    elif  row.Image_Label == 'laptop':
        folder = 'laptop'
        img_path = img_path + r'\laptop'
        shutil.copy(row.filename,img_path)
    
    elif  row.Image_Label == 'tabby':
        folder = 'tabby'
        img_path = img_path + r'\tabby'
        shutil.copy(row.filename,img_path)
        
    elif  row.Image_Label == 'violin':
        folder = 'violin'
        img_path = img_path + r'\violin'
        shutil.copy(row.filename,img_path)
        
    return folder
```

```python
df['folders'] = df.apply(copy_image_to_dir, axis=1)
```
