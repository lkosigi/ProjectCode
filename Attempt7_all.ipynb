### Data Sampling - downsample
#### This is done primarily because of the class imbalance in the data, the number of no's are much higher than yes, when such data is used the classifier is unsatisfactory. 
#### DownSampling helps with class imbalance and creates a smaller subset of data of data which is easier to handle given system limitations

```python
df_merged.is_exciting.value_counts()
```

```python
# clearly there is a class imbalance, to overcome this problem we downsample
#Ref: https://elitedatascience.com/imbalanced-classes
#Down-sample Majority Class

# Separate majority and minority classes
df_majority = df_merged[df_merged.is_exciting=='f']
df_minority = df_merged[df_merged.is_exciting=='t']
```

```python
# Downsample majority class
from sklearn.utils import resample
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=36710,     # to match minority class
                                 random_state=200) # reproducible results
```

```python
# Combine minority class with downsampled majority class
df = pd.concat([df_majority_downsampled, df_minority])
```

```python
# Display new class counts
df.is_exciting.value_counts()
```

### Data audit

```python
def data_audit(df, name):
    d=df.dtypes[df.dtypes!='object'].index.values
    df[d]=df[d].astype('float64')
    
    mean=DataFrame({'mean':df[d].mean()})
    std_dev=DataFrame({'std_dev':df[d].std()})
    missing= DataFrame({'missing':df[d].isnull().sum()})
    obs=DataFrame({'obs':np.repeat(df[d].shape[0],len(d))},index=d)
    missing_perc=DataFrame({'missing_perc':df[d].isnull().sum()/df[d].shape[0]})
    minimum=DataFrame({'min':df[d].min()})
    maximum=DataFrame({'max':df[d].max()})
    unique=DataFrame({'unique':df[d].apply(lambda x:len(x.unique()),axis=0)})
    q5=DataFrame({'q5':df[d].apply(lambda x:x.dropna().quantile(0.05))})
    q10=DataFrame({'q10':df[d].apply(lambda x:x.dropna().quantile(0.10))})
    q25=DataFrame({'q25':df[d].apply(lambda x:x.dropna().quantile(0.25))})
    q50=DataFrame({'q50':df[d].apply(lambda x:x.dropna().quantile(0.50))})
    q75=DataFrame({'q75':df[d].apply(lambda x:x.dropna().quantile(0.75))})
    q85=DataFrame({'q85':df[d].apply(lambda x:x.dropna().quantile(0.85))})
    q95=DataFrame({'q95':df[d].apply(lambda x:x.dropna().quantile(0.95))})
    q99=DataFrame({'q99':df[d].apply(lambda x:x.dropna().quantile(0.99))})
    DQ=pd.concat([mean,std_dev,obs,missing,missing_perc,minimum,maximum,unique,q5,q10,q25,q50,q75,q85,q95,q99],axis=1)
    
    c=df.dtypes[df.dtypes=='object'].index.values
    Mean=DataFrame({'mean':np.repeat('Not Applicable',len(c))},index=c)
    Std_Dev=DataFrame({'std_dev':np.repeat('Not Applicable',len(c))},index=c)
    Missing=DataFrame({'missing':df[c].isnull().sum()})
    Obs=DataFrame({'obs':np.repeat(df[d].shape[0],len(c))},index=c)
    Missing_perc=DataFrame({'missing_perc':df[c].isnull().sum()/df[c].shape[0]})
    Minimum=DataFrame({'min':np.repeat('Not Applicable',len(c))},index=c)
    Maximum=DataFrame({'max':np.repeat('Not Applicable',len(c))},index=c)
    Unique=DataFrame({'unique':df[c].apply(lambda x:len(x.unique()),axis=0)})
    Q5=DataFrame({'q5':np.repeat('Not Applicable',len(c))},index=c)
    Q10=DataFrame({'q10':np.repeat('Not Applicable',len(c))},index=c)
    Q25=DataFrame({'q25':np.repeat('Not Applicable',len(c))},index=c)
    Q50=DataFrame({'q50':np.repeat('Not Applicable',len(c))},index=c)
    Q75=DataFrame({'q75':np.repeat('Not Applicable',len(c))},index=c)
    Q85=DataFrame({'q85':np.repeat('Not Applicable',len(c))},index=c)
    Q95=DataFrame({'q95':np.repeat('Not Applicable',len(c))},index=c)
    Q99=DataFrame({'q99':np.repeat('Not Applicable',len(c))},index=c)
    dq=pd.concat([Mean,Std_Dev,Obs,Missing,Missing_perc,Minimum,Maximum,Unique,Q5,Q10,Q25,Q50,Q75,Q85,Q95,Q99],axis=1)

    DQ=pd.concat([DQ,dq])
    DQ.to_csv(name)

```

```python
### Text processing
import nltk
nltk.download('stopwords')
import string
from nltk.corpus import stopwords

```

```python
def text_process(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Remove all \r\n
    4. Remove all numbers
    5. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    no_punc = [char for char in mess if char not in string.punctuation + string.digits]

    # Join the characters again to form the string.
    no_punc = ''.join(no_punc)
    
    no_punc = no_punc.replace('\r\\n', '')
    
    # Now just remove any stopwords
    
    return (' '.join([word.lower() for word in no_punc.split() if word.lower() not in stopwords.words('english')]))   
     
```

### KFold cv for Logistic Regression

```python
#https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc_crossval.html
from sklearn.model_selection import KFold
from scipy import interp
```

```python
# https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc_crossval.html
# #############################################################################
# Classification and ROC analysis

# Run classifier with cross-validation and plot ROC curves
cv = KFold(5,True, 1)
classifier = linear_model.LogisticRegression()
tprs = []
aucs = []
folds = []
mean_fpr = np.linspace(0, 1, 100)
i = 0
```

```python
# https://stackoverflow.com/questions/36967666/transform-scipy-sparse-csr-to-pandas/36968458
t_df = pd.DataFrame(train_features.todense())
```

```python
for train, test in cv.split(t_df, y):
    X_train, X_test = t_df.values[train], t_df.values[test] 
    y_train, y_test = y.values[train], y.values[test]
    probas_ = classifier.fit(X_train, y_train).predict_proba(X_test)
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
    tprs.append(interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    folds.append(i+1)
    plt.plot(fpr, tpr, lw=1, alpha=0.3,
             label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))
    i += 1
plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
         label='Chance', alpha=.8)

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)
plt.plot(mean_fpr, mean_tpr, color='b',
         label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
         lw=2, alpha=.8)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                 label=r'$\pm$ 1 std. dev.')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for Logistic Regression')
plt.legend(loc="lower right")
plt.show()
```

```python
# https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc_crossval.html
# #############################################################################
# Classification and ROC analysis

# Run classifier with cross-validation and plot ROC curves
cv = KFold(5,True, 1)
classifier = RandomForestClassifier(n_estimators=300,oob_score=True,n_jobs=-1,random_state=400)
tprs = []
aucs = []
folds = []
mean_fpr = np.linspace(0, 1, 100)
i = 0

for train, test in cv.split(t_df, y):
    X_train, X_test = t_df.values[train], t_df.values[test] 
    y_train, y_test = y.values[train], y.values[test]
    probas_ = classifier.fit(X_train, y_train).predict_proba(X_test)
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
    tprs.append(interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    folds.append(i+1)
    plt.plot(fpr, tpr, lw=1, alpha=0.3,
             label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))
    i += 1
plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
         label='Chance', alpha=.8)

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)
plt.plot(mean_fpr, mean_tpr, color='b',
         label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
         lw=2, alpha=.8)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                 label=r'$\pm$ 1 std. dev.')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for RandomForest')
plt.legend(loc="lower right")
plt.show()
```
